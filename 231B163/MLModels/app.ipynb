```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score

# Load dataset
df = pd.read_csv("dataset1csv")

# Convert height/weight ranges to numeric averages
def convert_range(x):
    if isinstance(x, str) and "-" in x:
        a, b = x.split("-")
        return (float(a) + float(b)) / 2
    return float(x)

df['Height Range(CM)'] = df['Height Range(CM)'].apply(convert_range)
df['Weight Range(Kg)'] = df['Weight Range(Kg)'].apply(convert_range)

# Example target: classify consumption as High vs Low
df['Consumption_Class'] = df['Wh/km'].apply(lambda x: 1 if x > 20 else 0)  

# Drop irrelevant columns
X = df.drop(columns=['Trip ID','Starting Point','End Point','Consumption_Class'])
y = df['Consumption_Class']

# Encode categorical columns
for col in ['Weather','Wind Direction']:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# KNN model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# Evaluation
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy:", acc)
print("F1 Score:", f1)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_neighbors': range(3, 15),
    'metric': ['euclidean','manhattan']
}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)

print("Best Params:", grid.best_params_)
print("Best Accuracy:", grid.best_score_)
```
