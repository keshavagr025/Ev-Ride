```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score
from xgboost import XGBClassifier

# ===============================
# 1. Load Dataset
# ===============================
data = pd.read_excel("Keshav'sData.xlsx")

# ===============================
# 2. Target Variable → Ride Successful
# ===============================
# Ride is successful if paid with card, cash, or wallet
data["ride_successful"] = (
    (data["paid_using_card"] > 0) |
    (data["paid_cash"] > 0) |
    (data["paid_using_wallet"] > 0)
).astype(int)

y = data["ride_successful"]

# ===============================
# 3. Features
# ===============================
X = data.drop(columns=[
    "paid_using_card", "paid_cash", "paid_using_wallet", "ride_successful",
    "request_made_at", "request_accepted_at", "pickup_time", "ride_end_at",
    "user_id", "ride_id", "driver_id"
])

# Handle missing values
X["rating_given_to_driver"] = X["rating_given_to_driver"].fillna(X["rating_given_to_driver"].mean())

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# ===============================
# 4. XGBoost Model
# ===============================
xgb = XGBClassifier(eval_metric="logloss", random_state=42)
xgb.fit(X_train, y_train)

# Predictions
y_pred_xgb = xgb.predict(X_test)

# ===============================
# 5. Evaluation
# ===============================
acc = accuracy_score(y_test, y_pred_xgb)
f1 = f1_score(y_test, y_pred_xgb, average='weighted')

print("✅ Ride Success Prediction Accuracy:", acc)
print("✅ Ride Success Prediction F1 Score:", f1)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(xgb, X_scaled, y, cv=5, scoring='accuracy')
print("Cross-validation Accuracy:", scores.mean())
```
